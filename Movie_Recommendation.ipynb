```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movie['cast'] = movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

## TF-IDF VECTORIZATION

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
```

### Converting matrix to Array

```python
vectors = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Find the Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

## RECOMMEND FUNCTION

```python
def recommend(movie):
    movie = movie.lower()  # make search more robust
    movie_index = new_df[new_df['title'].str.lower() == movie].index[0]
    
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), 
                         reverse=True, key=lambda x: x[1])[1:6]

    print("\nTop Recommendations:\n")
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
def relevance(movie_idx1, movie_idx2):
    tags1 = set(new_df.iloc[movie_idx1].tags.split())
    tags2 = set(new_df.iloc[movie_idx2].tags.split())
    
    shared = tags1.intersection(tags2)
    return 1 if len(shared) > 0 else 0
```

```python
def precision_at_k(movie_idx, k=5):
    distances = similarity[movie_idx]
    movies_list = sorted(
        list(enumerate(distances)), key=lambda x: x[1], reverse=True
    )[1:k+1]
    
    relevant = 0
    for idx, _ in movies_list:
        if relevance(movie_idx, idx) == 1:
            relevant += 1
    
    return relevant / k
```

```python
def recall_at_k(movie_idx, k=5):
    tags1 = set(new_df.iloc[movie_idx].tags.split())
    
    # all movies in dataset that share tags
    total_relevant = 0
    for i in range(len(new_df)):
        if i != movie_idx:
            tags2 = set(new_df.iloc[i].tags.split())
            if len(tags1.intersection(tags2)) > 0:
                total_relevant += 1

    # top K recommended movies
    distances = similarity[movie_idx]
    movies_list = sorted(
        list(enumerate(distances)), key=lambda x: x[1], reverse=True
    )[1:k+1]

    found_relevant = 0
    for idx, _ in movies_list:
        if relevance(movie_idx, idx) == 1:
            found_relevant += 1

    return found_relevant / total_relevant if total_relevant > 0 else 0
```

```python
def evaluate_system(k=5):
    precisions = []
    recalls = []
    
    for i in range(0, 200):       # evaluate on 200 movies for speed
        precisions.append(precision_at_k(i, k))
        recalls.append(recall_at_k(i, k))
    
    print("Precision@{}: {:.4f}".format(k, np.mean(precisions)))
    print("Recall@{}: {:.4f}".format(k, np.mean(recalls)))
```

```python
evaluate_system(k=5)
```

```python
def evaluate_similarity(movie_name):
    movie_index = new_df[new_df['title'] == movie_name].index[0]
    distances = similarity[movie_index]
    recommended = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    scores = [x[1] for x in recommended]
    return {
        "average_similarity": np.mean(scores),
        "min_similarity": np.min(scores),
        "max_similarity": np.max(scores),
        "all_scores": scores
    }
```

```python
print(evaluate_similarity("Avatar"))
```
